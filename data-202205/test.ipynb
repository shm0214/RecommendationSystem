```python
import pandas as pd
```

```python
df = pd.read_table('./itemAttribute.txt', sep='|')
df.columns = ['item', 'attr1', 'attr2']
```

```python
df
```

```python
df['attr'] = df.apply(lambda x: sorted([x['attr1'], x['attr2']]), axis=1)
```

```python
df
```

```python
df1 = df.drop(['item', 'attr1', 'attr2'], axis=1)
```

```python
df1
```

```python
ls = list(df['attr'])
```

```python
ls
```

```python
ls1 = []
for a, b in ls:
    if a == 'None':
        a = -1
    if b == 'None':
        b = -1
    a, b = int(a), int(b)
    a, b = (a, b) if a < b else (b, a)
    s = '{},{}'.format(a, b)
    ls1.append(s)
```

```python
ls1
```

```python
dict = {}
for key in ls1:
    dict[key] = dict.get(key, 0) + 1
```
